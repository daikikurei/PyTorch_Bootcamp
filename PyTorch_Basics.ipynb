```python
x = torch.rand(5,3) #ダミーデータ #0～1の乱数
print(x)
```

```python
y = torch.rand(5,3) #ダミーデータ #0～1の乱数
print(y)
```

```python
print(x + y) #テンソルの足し算ができる。#対応する要素同士の演算
```

```python
result = torch.empty(5,3) #空のテンソル
torch.add(x,y, out=result) #出力の結果をresultに代入
```

```python
y.add_(x) #yにxを足す
print(y)
```

```python
print(y[:,1]) #[行は全部取得, 列は1列目だけ取得] #「:」スライス
```

```python
print(y[0,:]) #[行は0行目だけ取得, 列は全部取得]
```

```python
print(y[2:,1:]) #[行は2行目以降全部, 列は1列目以降全部]
```

```python
x = torch.randn(4, 4)
y = x.view(16) #16個の要素にする
z = x.view(-1, 8) # -1:自動計算 #列が8の場合に、自動的に行が2となる。
print(x.size(),y.size(),z.size())
```

```python
x = torch.randn(1) #テンソルで表現された値はテンソルオブジェクトで返ってくる。
print(x)
print(x.item()) #数値そのものが返ってくる。 #numpy等で使える
```

```python
a = torch.ones(3) #tensor型
print(a)
```

```python
b = a.numpy() #numpy型へ変換
print(b)
```

```python
a.add_(3) #aとbはメモリを共有しているので、aの値を変更するとbも変更される。
print(a)
print(b)
```

```python
a = np.ones(3) #nuympyのデータ列
print(a)
```

```python
b = torch.from_numpy(a) #tensor型へ変換
print(b)
```

```python
a = torch.arange(1, 4, 1, dtype=torch.float) #ダミーのデータ列 
print(a)
```

```python
a.mean() #tensor型の平均
```

```python
a.sum() #tensor型の総和
```

```python
print(a.max()) #最大
print(a.min()) #最小
```

```python
a = torch.tensor([1,2,3], dtype=torch.float)
b = torch.tensor([4,5,6], dtype=torch.float)

print(torch.add(a,b).sum()) #足して、その総和をとる
```

```python
import math
a = torch.tensor([math.radians(30), math.radians(60), math.radians(90)])
```

```python
#三角関数
print(torch.sin(a))
print(torch.cos(a))
print(torch.tan(a))
print(torch.asin(a))
print(torch.acos(a))
print(torch.tanh(a))
```

```python
a = torch.tensor([[1,2,3],[4,5,6]],dtype=torch.float)
b = torch.tensor([[1,2],[3,4],[5,6]],dtype=torch.float)
```

```python
torch.mm(a,b) #行列の掛け算
```

```python
print(a@b) #行列の掛け算
```

```python
x = torch.ones(2,2, requires_grad = True) #勾配を計算したい場合はrequires_grad = Trueが必要
```

```python
y = x + 2
```

```python
z = (y**2).mean()
```

```python
z.backward()
print(x.grad)
```
