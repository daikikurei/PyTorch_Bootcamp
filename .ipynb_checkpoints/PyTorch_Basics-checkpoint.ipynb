```python
import torch
import numpy as np
```

```python
x = torch.empty(5,3) #空の行列(0になる事は保証されていない) #5行3列
print(x)
```

```python
x = torch.rand(5,3) #0～1の乱数
print(x)
```

```python
torch.zeros(5,3, dtype=torch.long) #0
```

```python
torch.tensor([1.23,3]) #べた書きで書いてtensorのデータ型を作る事もできる
```

```python
x = x.new_ones(5,3, dtype=torch.double)
print(x)
```

```python
x = torch.randn_like(x,dtype=torch.float) #randn:正規分布 #like:5行3列の形をそのまま継承 #中身をfloatに
print(x)
```

```python
x = torch.arange(0,19,2) #0から18までを2個とばし
print(x)
```

```python
x = torch.linspace(0,3,4) #要素の数を指定 #0～3の範囲で4つの要素を自動生成
print(x)
```

```python
x = torch.FloatTensor([5,6,7]) #float型で確実に作る場合
print(x)
print(x.dtype)
```

```python
x = torch.tensor([5,6,7],dtype=torch.int) #このような形でもデータ型を指定できる。
```

```python
print(x.dtype)
```

```python
torch.manual_seed(42) #乱数を発生させるアルゴリズムのキー #これを指定しておくと、(同じセル内では)正規分布等を発生する時に毎回同じ結果を得られる
x = torch.rand(2,3)
print(x)
```

```python
x = torch.rand(2,3)
print(x)
```

```python
x.dtype #データ型を取得
```

```python
x.size() #マトリクスのサイズ #何行何列か
```

```python
x.shape #マトリクスのサイズ #何行何列か
```

```python
x = torch.rand(5,3) #ダミーデータ #0～1の乱数
print(x)
```

```python
y = torch.rand(5,3) #ダミーデータ #0～1の乱数
print(y)
```

```python
print(x + y) #テンソルの足し算ができる。#対応する要素同士の演算
```

```python
result = torch.empty(5,3) #空のテンソル
torch.add(x,y, out=result) #出力の結果をresultに代入
```

```python
y.add_(x) #yにxを足す
print(y)
```

```python
print(y[:,1]) #[行は全部取得, 列は1列目だけ取得] #「:」スライス
```

```python
print(y[0,:]) #[行は0行目だけ取得, 列は全部取得]
```

```python
print(y[2:,1:]) #[行は2行目以降全部, 列は1列目以降全部]
```

```python
x = torch.randn(4, 4)
y = x.view(16) #16個の要素にする
z = x.view(-1, 8) # -1:自動計算 #列が8の場合に、自動的に行が2となる。
print(x.size(),y.size(),z.size())
```

```python
x = torch.randn(1) #テンソルで表現された値はテンソルオブジェクトで返ってくる。
print(x)
print(x.item()) #数値そのものが返ってくる。
```
